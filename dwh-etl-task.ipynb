### <b>What is ETL process?</b>
ETL is a process in data warehousing which stands for Extract, Transform and Load. It’s a process of extracting huge amounts of data from a variety of sources and transforming the extracted data into a well organized and readable format via techniques like data aggregation and data normalization and finally at last loading it into a storage system like database, data warehouses.

### Applying Transformations on the Datasets

```python
#Slicing Data to Remove Unecessary Columns
covidStats = covid_data[['hash','date', 'states', 'positive', 'negative', 'hospitalized', 'totalTestResults', 'death']].set_index('hash')
covidStats.head(3)
```

<b> Ecommerce Store Dataset </b>

```python
# Order Details of an Ecommerce Store
order_details
```

```python
# Orders Data
orders
```

```python
# Merge the two DataFrames on the "order id" column
orderData = pd.merge(order_details, orders, how='inner', on='Order ID')

# Sort the merged DataFrame by the "date" column in descending order
orderData = orderData.sort_values("Order Date", ascending=False)

# Resetting the index after sorting
orderData = orderData.reset_index(drop=True)

# Calculate a new column "Average Order Value" by Dividing "amount" and "quantity"
orderData['AOV'] = orderData['Amount'] / orderData['Quantity'].astype(int)

# Display the resulting DataFrame
print(orderData)
```
